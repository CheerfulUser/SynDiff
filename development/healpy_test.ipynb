```python
import matplotlib.pyplot as plt
%matplotlib notebook

import numpy as np
import healpy as hp
```

```python
NSIDE = 320
print(
    "Approximate resolution at NSIDE {} is {:.2} deg".format(
        NSIDE, hp.nside2resol(NSIDE, arcmin=True) / 60
    )
)
```

```python
NPIX = hp.nside2npix(NSIDE)
print(NPIX)
```

```python
m = np.arange(NPIX)
hp.mollview(m, title="Mollview image RING")
hp.graticule()
```

```python
vec = hp.ang2vec(np.pi / 2, np.pi * 3 / 4)
print(vec)
```

```python
ipix_disc = hp.query_disc(nside=320, vec=vec, radius=np.radians(10))
m = np.arange(NPIX)
m[ipix_disc] = m.max()
hp.mollview(m, title="Mollview image RING")
```
