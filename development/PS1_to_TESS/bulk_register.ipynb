```python
import sys 
sys.path.append('../../src/')
from register_pix import compute_coverage
from glob import glob 
import numpy as np
```

```python
files = np.array(glob('../data/ps1_skycell/*.satcor*'))
tess_file = 'data/tess2019359002923-s0020-1-1-0165-s_ffic.fits'
```

```python
cell = np.array([f.split('.stk')[0] + '.stk.' for f in files])
```

```python
good = []
for i in range(len(cell)):
    if sum(cell[i] == cell) == 4:
        good += [cell[i]]
good = list(set(good))
good = np.array([g + 'r.unconv.satcor.fits'for g in good])
```

```python
done = glob('bulk_test_buffer/*.npy')
done = np.array([f.split('.npy')[0].split('/')[-1] for f in done])

todo = []
cell2 = np.array([f.split('.stk')[0].split('/')[-1] for f in good])
for i in range(len(cell2)):
    if (cell2[i] == done).any():
        print('done ', cell2[i])
    else:
        #print('not done')
        todo += [i]
todo = np.array(todo)
todo = good[todo]
        
for file in todo:
    compute_coverage(tess_file,file,verbose=True,save_path='bulk_test_buffer')
```
